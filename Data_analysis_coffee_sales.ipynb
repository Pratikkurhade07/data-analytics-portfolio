```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
customers = pd.read_excel(r'C:\Users\Prati\Desktop\coffee sales\\Raw Data.xlsx',sheet_name="customers")
orders=pd.read_excel(r'C:\Users\Prati\Desktop\coffee sales\\Raw Data.xlsx',sheet_name="orders")
products=pd.read_excel(r'C:\Users\Prati\Desktop\coffee sales\\Raw Data.xlsx',sheet_name="products")
products.head()
```

```python
customers.head()
```

```python
orders.head()
```

```python
data=pd.merge(customers,orders,on='Customer ID')
data=pd.merge(data,products, on='Product ID')
data.head()
```

```python
data.info()
```

```python
data=data.drop(columns=['Phone Number',"Email_x",'Customer Name_y','Email_y','Country_y','Coffee Type_x','Roast Type_x','Size_x','Unit Price_x','Sales'])
data=data.dropna(how='all')
data.info()
```

```python
data.isnull().values.sum().sum()
```

```python
plt.figure(figsize = (5,5))
sns.heatmap(data[['Size_y','Unit Price_y','Profit']].corr(), cmap="YlGnBu",
            annot=True)
```

```python
#Profit and size of cofee

x=data.Size_y
y=data.Profit

# Create scatter plot
plt.scatter(x, y,color='green',edgecolors="Black")
plt.title('Profit vs coffee size')
plt.xlabel('Coffee size')
plt.ylabel('Profit')

# Show plot
plt.show()
```

```python
#country vs quantity

x1=data["Country_x"].unique()
x2=list(x1)
x2.sort()
y2=data.groupby(["Country_x"])["Quantity"].sum()
print(y2)

# Create scatter plot
plt.bar(x2, y2,color="blue")
plt.title('Countries vs Quantity Ordered')
plt.xlabel('Countries')
plt.ylabel('Quantity Ordered')

# Show plot
plt.show()
```

```python
x1=data["Country_x"].unique()
x2=list(x1)
x2.sort()
y1=data.groupby(["Country_x"])["Profit"].mean()
print(y1)
y3=data.groupby(["Country_x"])["Size_y"].mean()
print(y2)


fig,ax1 = plt.subplots()
ax2 =ax1.twinx()
ax2.plot(y3,"o-",color="Orange",label='Average Bean Size')
ax1.set_xticklabels( x2 ,size=8)
ax1.set_xlabel('Countries') 
ax1.set_ylabel('Average Profits') 
ax2.set_ylabel("Average Bean Size Product") 

ax1.bar(x2,y1,color="blue", edgecolor="black")
plt.legend(loc='best')
plt.title("Average Profit and  Coffee bean size ordered by Countries")
plt.show()
```

```python
Ct=data["Coffee Type_y"].unique()
x3=list(Ct)
x3.sort()
y4=data.groupby(["Coffee Type_y"])["Profit"].sum()
print(y1)
y5=data.groupby(["Coffee Type_y"])["Quantity"].sum()



fig,ax1 = plt.subplots()
ax2 =ax1.twinx()
ax2.plot(y5,"o-",color="orange",label='Quantity Ordered')
ax1.set_xticklabels( x3,rotation="vertical" ,size=8)
ax1.set_xlabel('Type of Coffee') 
ax1.set_ylabel('Total Profit') 
ax2.set_ylabel("Total Quantity Ordered") 

ax1.bar(x3,y4,color="blue", edgecolor="black")
plt.legend(loc='best')
plt.title("Type of coffee bean vs profit and quantity")
plt.show()
```

```python
LC=data["Loyalty Card"].unique()
LC1=list(LC)
LC1.sort()
y7=data.groupby(["Loyalty Card"])["Loyalty Card"].count()
print(y7)

# Create scatter plot
plt.pie(y7,labels=LC1, autopct='%1.1f%%')
plt.title('Transaction of order based with/out Loyalty Card ')


# Show plot
plt.show()
```

```python
LC=data["Loyalty Card"].unique()
LC1=list(LC)
LC1.sort()
y8=data.groupby(["Loyalty Card"])["Quantity"].sum()

y9=data.groupby(["Loyalty Card"])["Profit"].sum()
print(y8)
print(y9)


fig,ax1 = plt.subplots()
ax2 =ax1.twinx()
ax2.plot(y9,"o-",color="orange",label='Profit')
ax1.set_xticklabels( LC1,rotation="vertical" ,size=12)
ax1.set_xlabel('Loyalty Card') 
ax1.set_ylabel('Total Quantity Ordered') 
ax2.set_ylabel("Profit") 

ax1.bar(LC1,y8,color="blue", edgecolor="black")
plt.legend(loc='best')
plt.title("Type of coffee bean vs profit and quantity")
plt.show()
```
